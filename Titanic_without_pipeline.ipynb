```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df1.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
from re import X
#Step1 test-train-split
X_train, X_test, y_train, y_test= train_test_split(df1.drop(columns=['Survived']),df1['Survived'],test_size=0.2,random_state=42)
X_train.shape, X_test.shape
```

```python
X_train.head(2)
```

```python
y_train.head()
```

```python
"missing values"
df1.isnull().sum()
```

```python
#Age and fare missing values
#fit and transform
si_age= SimpleImputer()
si_fare= SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_fare = si_fare.fit_transform(X_train[['Fare']])

X_test_age = si_age.transform(X_test[['Age']])
X_test_fare = si_fare.transform(X_test[['Fare']])
```

```python
#onehot encoding on sex and embarked
ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train[['Embarked']])

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test[['Embarked']])
```

```python
X_train.head(2)
```

```python
X_train_sex
```

```python
X_test_embarked
```

```python
#4 arrays age,sex,fair,embarked concatinate
X_train_rem = X_train.drop(columns=['Age','Sex','Embarked','Fare'])
X_test_rem = X_test.drop(columns=['Age','Sex','Embarked','Fare'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_sex,X_train_embarked,X_train_age,X_train_fare),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_sex,X_test_embarked,X_test_age,X_test_fare),axis=1)
```

```python
X_train_transformed.shape
```

```python
clf=  DecisionTreeClassifier()
clf.fit(X_train_transformed,y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
