```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = keras.datasets.fashion_mnist
```

```python
(train_images,train_labels),(test_images,test_labels) = data.load_data()
```

```python
class_names= ['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot']
train_images=train_images/255.0
test_images=test_images/255.0
```

```python
plt.imshow(train_images[70])
plt.show()
```

```python
plt.imshow(train_images[60],cmap=plt.cm.binary)
plt.show()
```

```python
#now we will make a sexy ahh NEURAL NETWORK model which will tell us which image is which label
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation="relu"),
    keras.layers.Dense(10,activation="softmax")
    ])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
model.fit(train_images,train_labels, epochs=10)
#this whole block will train the code the activation and optimizers with matrix read about it
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("tested acc",test_acc)
#testing the model which i made rn
```

```python
prediction = model.predict(test_images)
print(class_names[np.argmax(prediction[0])])

print("tested acc",test_acc)
#real prediction time
```
